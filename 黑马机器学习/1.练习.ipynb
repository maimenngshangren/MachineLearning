## 1.2 添加样式

```python
# 解决中文不显示问题
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False

import random
# 准备数据
x = range(60)
y = [random.uniform(15,18) for i in x]     # 列表生成式
# 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 绘制图像
plt.plot(x,y)
# 添加X，Y轴刻度
x_ticks_label=['11点{}分'.format(i) for i in x]
y_ticks=range(40)
# 修改X，Y轴坐标显示
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])
# 添加网格显示
plt.grid(True, linestyle='--', alpha=0.5)
# 添加X，Y轴信息及标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('温度变化', fontsize=20)

# 图像保存
plt.savefig('test.png')

# 图像显示
plt.show()
```

## 1.3一个坐标系绘制多个图像

```python
# 解决中文不显示问题
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False

import random

# 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]     # 列表生成式
y_beijing = [random.uniform(12,15) for i in x]

# 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 绘制图像
plt.plot(x,y_shanghai, color='r', linestyle='-', label='上海')
plt.plot(x,y_beijing, color='b', linestyle='--', label='北京')

# 添加图例
plt.legend(loc='best')

# 添加X，Y轴刻度
x_ticks_label=['11点{}分'.format(i) for i in x]
y_ticks=range(40)

# 修改X，Y轴坐标显示
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 添加网格显示
plt.grid(True, linestyle='--', alpha=0.5)

# 添加X，Y轴信息及标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('温度变化', fontsize=20)

# 图像保存
plt.savefig('test1.png')

# 图像显示
plt.show()
```

## 1.4 多个坐标系绘制多个图像

```python
# 解决中文不显示问题
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False

import random

# 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]     # 列表生成式
y_beijing = [random.uniform(12,15) for i in x]

# 创建画布
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 绘制图像
axes[0].plot(x,y_shanghai, color='r', linestyle='-', label='上海')
axes[1].plot(x,y_beijing, color='b', linestyle='--', label='北京')

# 添加X，Y轴刻度
x_ticks_label=['11点{}分'.format(i) for i in x]
y_ticks=range(40)

# 修改X，Y轴坐标显示
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])

# 添加网格显示
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)

# 添加X，Y轴信息及标题
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('温度变化', fontsize=20)
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('温度变化', fontsize=20)

# 显示图例
axes[0].legend(loc=0)
axes[1].legend(loc=0)

# 图像保存
plt.savefig('test2.png')

# 图像显示
plt.show()
```
