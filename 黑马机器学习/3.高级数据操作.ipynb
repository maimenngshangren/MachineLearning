## 2.1 数据分析

## 数据相关性

```python
data_cor=data.loc[:,['RPM', 'AGE', 'SALARY_MILLIONS', 'ORB', 'DRB', 'TRB',
                       'AST', 'STL', 'BLK', 'TOV', 'PF', 'POINTS', 'GP', 'MPG', 'ORPM', 'DRPM']]
data_cor.head()
```

```python
# 获取两列的相关性
corr=data_cor.corr()
corr.head()
```

```python
plt.figure(figsize=(20,8),dpi=100)
sns.heatmap(corr,square=True,linewidths=0.1,annot=True)
```

## 基本数据排名分析

```python
data.loc[:,['PLAYER','RPM',"AGE"]].sort_values(by='RPM',ascending=False)
```

## 数据可视化

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))

sns.displot(data['SALARY_MILLIONS'])
plt.ylabel('SALARY_MILLIONS')

sns.displot(data['RPM'])
plt.ylabel('RPM')

sns.displot(data['AGE'])
plt.ylabel('AGE')
```

```python
sns.jointplot([data.AGE, data.SALARY_MILLIONS])
```

```python
multi_data=data.loc[:,['RPM','AGE','POINTS','SALARY_MILLIONS']]
multi_data.head()
```

```python
sns.pairplot(multi_data)
```

```python
def age_cut(df):
    if df.AGE < 24:
        return 'young'
    elif df.AGE > 30:
        return 'old'
    else:
        return ('best')
data['age_cut']=data.apply(lambda x:age_cut(x),axis=1)
data['cut']=1
```

```python
data.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,10),dpi=100)
plt.title('RPM and Salary',size=15)

x1=data.loc[data.age_cut=='best'].SALARY_MILLIONS
y1=data.loc[data.age_cut=='best'].RPM
plt.plot(x1,y1,'.')

x2=data.loc[data.age_cut=='young'].SALARY_MILLIONS
y2=data.loc[data.age_cut=='young'].RPM
plt.plot(x2,y2,'.')

x3=data.loc[data.age_cut=='old'].SALARY_MILLIONS
y3=data.loc[data.age_cut=='old'].RPM

plt.plot(x3,y3,'.')
```

```python
multi_data2=data.loc[:,['RPM','SALARY_MILLIONS','TRB','AST','STL','BLK','age_cut']]
sns.pairplot(multi_data2,hue='age_cut')
```

```python
# 分组
#data.groupby(by='age_cut')['SALARY_MILLIONS'].mean()
data.groupby(by='age_cut').agg({'SALARY_MILLIONS':np.mean})
```

```python
data_team=data.groupby(by='TEAM').agg({'SALARY_MILLIONS':np.mean})
data_team.sort_values(by='SALARY_MILLIONS',ascending=False).head(10)
```

```python
data_rpm=data.groupby(by=['TEAM','age_cut']).agg({'SALARY_MILLIONS':np.mean,'RPM':np.max,'PLAYER':np.size})
```

```python
data_rpm.sort_values(by=['PLAYER','RPM'],ascending=False).head(10)
```

## 绘制箱线图和小提琴图

```python
# 绘制箱线图
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))

data_team2=data[data.TEAM.isin(['GS','CLE','SA','LAC','OKC','UTAH','CHA','TOR','NO','BOS'])]

plt.subplot(3,1,1)
sns.boxplot(x='TEAM',y='SALARY_MILLIONS',data=data_team2)

plt.subplot(3,1,2)
sns.boxplot(x='TEAM',y='AGE',data=data_team2)

plt.subplot(3,1,3)
sns.boxplot(x='TEAM',y='MPG',data=data_team2)
```

```python
# 绘制小提琴图
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))

data_team2=data[data.TEAM.isin(['GS','CLE','SA','LAC','OKC','UTAH','CHA','TOR','NO','BOS'])]

plt.subplot(3,1,1)
sns.violinplot(x='TEAM',y='SALARY_MILLIONS',data=data_team2)

plt.subplot(3,1,2)
sns.violinplot(x='TEAM',y='AGE',data=data_team2)

plt.subplot(3,1,3)
sns.violinplot(x='TEAM',y='MPG',data=data_team2)
```
